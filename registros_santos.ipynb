```python
registros = pd.read_csv('http://www.dezenovevinte.net/registros_santos.csv')
```

```python
# Bairro de Ramos latitude and longitude values
latitude = -22.851693  
longitude = -43.255570
```

```python
!conda install -c conda-forge folium=0.5.0 --yes
import folium

print('Folium installed and imported!')
```

```python
# generate Ramos map and display it
ramos_map = folium.Map(location=[latitude, longitude], zoom_start=11)

# display the map of RJ
ramos_map
```

```python
# instantiate a feature group for the registros in the dataframe
santos = folium.map.FeatureGroup()

# loop through the 100 crimes and add each to the incidents feature group
for lat, lng, in zip(registros.Latitude, registros.Longitude):
    santos.add_child(
        folium.features.CircleMarker(
            [lat, lng],
            radius=5, # define how big you want the circle markers to be
            color='yellow',
            fill=True,
            fill_color='blue',
            fill_opacity=0.6
        )
    )
    
# add pop-up text to each marker on the map
latitudes = list(registros.Latitude)
longitudes = list(registros.Longitude)
labels = list(registros.Iconografia)

for lat, lng, label in zip(latitudes, longitudes, labels):
    folium.Marker([lat, lng], popup=label).add_to(ramos_map)   
    
# add registros to map
ramos_map.add_child(santos)
```
